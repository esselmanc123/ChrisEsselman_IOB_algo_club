```python
import random
import numpy as np
import math
from scipy.stats import binom
# Initialize Probabilities and number of flips
totalRounds = 100
coinFlipsPerRound = 100
Coin1_head_prob = 0.8
Coin2_head_prob = 0.2
```

```python
# Initialize vector and matrix. Vector will hold the coin used for each round. Matrix will hold the coin flips
coin_per_round = [0 for i in range(totalRounds)]
flips_matrix = [[0 for i in range(coinFlipsPerRound)] for j in range(totalRounds)]

# Do the coin flipping
for i in range(totalRounds):
    coin_per_round[i] = random.randint(1,2)
    if coin_per_round[i] == 1:
        flips_matrix[i] = random.choices([0,1], weights=[1-Coin1_head_prob, Coin1_head_prob], k=coinFlipsPerRound)
        
    else:
        flips_matrix[i] = random.choices([0,1], weights=[1-Coin2_head_prob, Coin2_head_prob], k=coinFlipsPerRound)
        
print(coin_per_round)
print(flips_matrix)
        
    
```

```python
# Do the EM algorithm
estimate_coin1_prob = random.uniform(0,1)
estimate_coin2_prob = 1-estimate_coin1_prob

for s in range(10000):
    # Make two lists that will hold the calculations
    #coin1_calc = [0 for i in range(totalRounds)]
    #coin2_calc = [0 for i in range(totalRounds)]
    coin1_calc = 0
    coin2_calc = 0
    
    # The variables where we will sum
    heads_coin1 = 0
    tails_coin1 = 0
    heads_coin2 = 0
    tails_coin2 = 0
    
    # For each flips matrix row
    for i in range(totalRounds):
        # Caclulate likelihood
        # First input is number of heads, second is number of trials, third in prob of head
        coin1_calc = binom.logpmf(sum(flips_matrix[i]),coinFlipsPerRound,estimate_coin1_prob)
        coin2_calc = binom.logpmf(sum(flips_matrix[i]),coinFlipsPerRound,estimate_coin2_prob)
    
        # Next calculate the probability
        coin1_calc_hold = coin1_calc/(coin1_calc + coin2_calc)
        coin2_calc = coin2_calc/(coin2_calc + coin1_calc)
        coin1_calc = coin1_calc_hold
    
        heads_coin1 = heads_coin1 + (sum(flips_matrix[i])*coin1_calc)
        tails_coin1 = tails_coin1 + ((coinFlipsPerRound-sum(flips_matrix[i]))*coin1_calc)
        heads_coin2 = heads_coin2 + (sum(flips_matrix[i])*coin2_calc)
        tails_coin2 = tails_coin2 + ((coinFlipsPerRound-sum(flips_matrix[i]))*coin2_calc)


    estimate_coin1_prob = heads_coin1/(heads_coin1+tails_coin1)
    estimate_coin2_prob = heads_coin2/(heads_coin2+tails_coin2)




print(estimate_coin1_prob)
print(estimate_coin2_prob)
```
