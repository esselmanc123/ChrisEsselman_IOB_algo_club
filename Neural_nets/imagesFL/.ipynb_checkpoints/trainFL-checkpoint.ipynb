```python
import numpy as np
import scipy.io

# Load the .mat file (replace filename with your actual .mat filename)
mat = scipy.io.loadmat('train_imagesFL.mat')

# Extract variables (MATLAB structs stored as dict in Python, fields get extra dimension sometimes)
x_train = mat['x_train']  # shape: (50000, 28, 28)
y_train = mat['y_train']  # shape: (50000, 1)
x_test = mat['x_test']    # shape: (1000, 28, 28)
y_test = mat['y_test']    # shape: (1000, 1)
print('Shapes:', x_train.shape, y_train.shape, x_test.shape, y_test.shape)
```

```python
y_train = y_train.squeeze()-1
y_test = y_test.squeeze()-1
print('Shapes2',y_train.shape,y_test.shape)
```

```python
# Making a model
from tensorflow.keras import Sequential
from keras.layers import Dense
from keras.layers import Flatten

# Create a sequential model
model = Sequential()

# Make a flatten layer that we will input 
model.add(Flatten(input_shape=(50,50)))
#Input and hidden layer
model.add(Dense(200, activation="relu"))
model.add(Dense(200, activation="relu"))
# Add final 1 neuron layer
#model.add(Dense(1))

# Maybe try a different output layer
model.add(Dense(10,activation="softmax"))

model.summary()
```

```python
# Hyperparameters for the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=["accuracy"])
```

```python
# Training the model
history = model.fit(x_train,y_train,epochs=50,validation_data=(x_test,y_test),batch_size=32,verbose=1)
```

```python
# Inspecting the results
import pandas as pd
import matplotlib.pyplot as plt
pd.DataFrame(history.history).plot(figsize=(15,8))
plt.grid(True)
#plt.gca().set_ylim(0,1)
plt.show()
```

```python
y_prob = model.predict(x_test)
y_pred = y_prob.argmax(axis=-1)
print(y_pred)
```

```python
import seaborn as sns
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_classification
y_prediction = model.predict(x_test)
y_prediction = np.argmax (y_prediction, axis = 1)
#Create confusion matrix and normalizes it over predicted (columns)
cm = confusion_matrix(y_test, y_prediction)
print("Confusion Matrix:\n", cm)
plt.figure(figsize=(8, 6))

# Use seaborn to create a heatmap of the confusion matrix
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)

# Add labels and title for clarity
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix for Classification Model')

# Display the plot
plt.show()
```
